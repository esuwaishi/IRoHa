```python
import numpy as np
from scipy.fftpack import fft
import matplotlib.pyplot as plt

# parameters
N = 2**22 # data number
#N = 5120*200
#dt = 0.0001 # data step [s]
dt = 1/5120 * 1/2


t = np.arange(0, N*dt, dt) # time
freq = np.linspace(0, 1.0/dt, N) # frequency step


y = np.zeros(N)*dt
#t0 = 10000
t0 = int(5120*2)
y[::t0] =1

yf = fft(y)/(N/2) # 離散フーリエ変換&規格化

plt.figure(2)
plt.subplot(211)
plt.plot(t, y)
plt.xlim(0, 5.5)
plt.xticks(np.arange(0, 6, 1), ["0", "$T_0$", "$2T_0$", "$3T_0$", "$4T_0$", "$5T_0$"])
plt.xlabel("time")
plt.ylabel("amplitude")

plt.subplot(212)
plt.plot(freq, 5e3*np.abs(yf))
plt.xlim(0, 5.5)
#plt.ylim(0, 0.25)
plt.xlabel("frequency")
plt.ylabel("amplitude")
plt.xticks(np.arange(0, 6, 1), ["0", "$\omega_0$", "$2\omega_0$", "$3\omega_0$", "$4\omega_0$", "$5\omega_0$"])
plt.tight_layout()
#plt.savefig("single_bunch.pdf", transparent=True)
plt.show()
```

```python
import numpy as np
from scipy.fftpack import fft
import matplotlib.pyplot as plt

# parameters
N = 2**20 # data number
dt = 0.0001 # data step [s]

t = np.arange(0, N*dt, dt) # time
freq = np.linspace(0, 1.0/dt, N) # frequency step

y = np.zeros(N)*dt
t0 = 1000
y[::t0] =1

for j in range(1, 100):
    for i in range(j*10000 - 1000, j*10000, t0):
        y[i] = 0

yf = fft(y)/(N/2) # 離散フーリエ変換&規格化

plt.figure(2)
plt.subplot(211)
plt.plot(t, y)
plt.xlim(0, 6)
plt.xticks(np.arange(0, 6, 1), ["0", "$T_0$", "$2T_0$", "$3T_0$", "$4T_0$", "$5T_0$"])
plt.xlabel("time")
plt.ylabel("amplitude")

plt.subplot(212)
plt.plot(freq, 5e3*np.abs(yf))
plt.xlim(0, 20)
plt.ylim(0, 1)
plt.xlabel("frequency$/\omega_0$")
plt.ylabel("amplitude")
#plt.xticks(np.arange(0, 6, 1), ["0", "$\omega_0$", "$2\omega_0$", "$3\omega_0$", "$4\omega_0$", "$5\omega_0$"])
plt.xticks(np.arange(0, 20, 1))
plt.tight_layout()
#plt.savefig("01")
plt.show()
```
