```python
import numpy as np
from scipy.fftpack import fft
import matplotlib.pyplot as plt

# parameters
N = 2**24 # data number
t0 = 2**14
omg0 = 2*np.pi*(1/t0)
omgs = omg0/10
dt = 1/t0
#dt = 0.0001 # data step [s]

t = np.arange(0, N*dt, dt) # time
freq = np.linspace(0, 1.0/dt, N) # frequency step

y = np.zeros(N)*dt
y[::t0] =1

yf = fft(y)/(N/2) # 離散フーリエ変換&規格化

plt.figure(2)
plt.subplot(211)
plt.plot(t, y)
plt.xlim(0, 5.5)
plt.xticks(np.arange(0, 6, 1), ["0", "$T_0$", "$2T_0$", "$3T_0$", "$4T_0$", "$5T_0$"])
plt.xlabel("time")
plt.ylabel("amplitude")

plt.subplot(212)
plt.plot(freq, 5e3*np.abs(yf))
plt.xlim(0, 5.5)
#plt.ylim(0, 0.25)
plt.xlabel("frequency")
plt.ylabel("amplitude")
plt.xticks(np.arange(0, 6, 1), ["0", "$\omega_0$", "$2\omega_0$", "$3\omega_0$", "$4\omega_0$", "$5\omega_0$"])
plt.tight_layout()
#plt.savefig("single_bunch.pdf", transparent=True)
plt.show()
```

```python
import numpy as np
from scipy.fftpack import fft
import matplotlib.pyplot as plt

# parameters
N = 2**24 # data number
#dt = 0.0001 # data step [s]
t0 = 2**14
omg0 = 2*np.pi*(1/t0)
omgs = omg0/10
dt = 1/t0

t = np.arange(0, N*dt, dt) # time
freq = np.linspace(0, 1.0/dt, N) # frequency step

y = np.zeros(N)*dt

tau = 1/(20*omgs)

for i in range(1, int(N/t0)):
    #tn = int(i*t0 - tau*np.cos(omgs*t0*i))
    tn = int(i*t0 + tau*np.cos(omgs*t0*i))
    y[tn] = 1

yf = fft(y)/(N/2) # 離散フーリエ変換&規格化

plt.figure(2)
plt.subplot(211)
plt.plot(t, y)
plt.xlim(0, 10)
#plt.xticks(np.arange(0, 6, 1), ["0", "$T_0$", "$2T_0$", "$3T_0$", "$4T_0$", "$5T_0$"])
plt.xticks(np.arange(0, 11, 1), ["0", "$T_0$", "$2T_0$", "$3T_0$", "$4T_0$", "$5T_0$","$6T_0$", "$7T_0$", "$8T_0$", "$9T_0$", "$10T_0$"])
plt.xlabel("time")
plt.ylabel("amplitude")

plt.subplot(212)
plt.plot(freq, 5e3*np.abs(yf))
plt.xlim(0.5, 1.5)
#plt.ylim(0, 0.5)
#plt.xticks(np.arange(0.5, 1.5, 0.1))
plt.xticks(np.arange(0.5, 1.5, 0.1), ["", "", "", "", "$\omega_0 - \omega_s$", "$\omega_0$","$\omega_0 + \omega_s$", "", "", ""])
plt.xlabel("frequency")
plt.ylabel("amplitude")
#plt.xticks(np.arange(0, 6, 1), ["0", "$\omega_0$", "$2\omega_0$", "$3\omega_0$", "$4\omega_0$", "$5\omega_0$"])
plt.tight_layout()
#plt.savefig("synchrotron_bunch.pdf", transparent=True)
plt.show()
```
