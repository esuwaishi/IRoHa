```python
import numpy as np
import matplotlib.pyplot as plt
import longitudinal as lde
```

```python
fig, ax = plt.subplots()

E0_e = 0.5109989461e6
E0_p = 938e6

Ek = np.linspace(1e3, 1e11, 50000000)

beta_e = np.sqrt(1 - 1/(1 + Ek/E0_e)**2)
beta_p = np.sqrt(1 - 1/(1 + Ek/E0_p)**2)

ax.semilogx(Ek, beta_e, label='electron')
ax.semilogx(Ek, beta_p, label='proton')
ax.set_xlabel('Kinetic energy')
ax.set_ylabel(r"Velocity $\beta\; (=v/c)$")
ax.set_xticks([1e3, 1e4, 1e5, 1e6, 1e7, 1e8, 1e9, 1e10, 1e11])
ax.set_xticklabels(["1 keV", "", "", "1 MeV", "", "", "1 GeV", "", ""])

ax.grid()
ax.legend().get_frame().set_alpha(0.6)
ax.legend(loc='lower right', fontsize='x-large', frameon=False)

fig.savefig("velocity.pdf", transparent=True)

plt.show()
```

```python
e_volt = 9.4e6
u0 = 1.76e6
sin_phi_s = u0/e_volt

mass = 0.5109989461e6
harm = 5120
frf = 508.887e6
frev = frf/harm
mcf = 3.2e-4
beam_energy = 4.0e9

gamma = beam_energy/mass
eta = mcf - 1/gamma/gamma
beta = np.sqrt(1 - 1/gamma/gamma)
cos_phi_s = np.sqrt(1 - (sin_phi_s)**2)
phi_s = np.arcsin(sin_phi_s)

fs = np.sqrt(harm*frev*frev*e_volt*eta*np.cos(phi_s) / (2*np.pi*beam_energy*beta*beta))
nus = fs/frev
inv_nus = frev/fs

print(fs/1e3)
print(nus)
print(inv_nus)
```
